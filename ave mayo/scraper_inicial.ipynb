```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada

#MODIFICAR URL
url = "https://www.jumbo.cl/limpieza"
headers = {"User-Agent": "Mozilla/5.0"} # Esto es para que no nos banee Garena

# Petición al servidor
respuesta = requests.get(url, headers=headers)

# Convertir HTML en objeto python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# Buscar los datos como títulos 


#MODIFICAR FIND_ALL
resultados = sopa.find_all('div', class_='flex flex-col')

# Mostrar en pantalla
for dato in resultados:
    print(f"Encontrado: {dato.text.strip()}")
```
