```python
from pymongo import MongoClient

# 1. Conectar al contenedor de Mongo
client = MongoClient('mongodb://mongodb:27017/')

# 2. Crear (o usar) la base de datos y la colección
db = client['TiendaBigData'] 
coleccion = db['SantaIsabelLimpiezaBano']  #Cambiado para productos de limpieza Jumbo

print("Conexión exitosa a MongoDB")
print(f"Base de datos: TiendaBigData")
print(f"Colección: SantaIsabelLimpiezaBano")
```

```python
import os
import time
import re
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# 1. Limpieza de procesos previos
os.system("pkill -9 chrome")
os.system("pkill -9 chromedriver")
print("Limpieza completada")

# 2. Configuración del navegador
options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1366,768")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = None
datos_finales = []
base_url = "https://www.santaisabel.cl/lacteos-huevos-y-congelados"

print("Iniciando scraping de Lacteos (25 paginas)...")

try:
    driver = webdriver.Chrome(options=options)
    
    for pagina in range(1, 26):
        url = f"{base_url}?page={pagina}" if pagina > 1 else base_url
        print(f"\n{'='*20} PAGINA {pagina}/25 {'='*20}")
        print(f"URL: {url}")
        
        driver.get(url)
        time.sleep(6) # Espera para carga de JS
        
        # Localizar contenedores de productos
        bloques = driver.find_elements(By.CSS_SELECTOR, 
            "div[class*='product'], div[class*='Product'], div[class*='item'], div[class*='tile'], article"
        )
        
        print(f"Bloques encontrados en esta pagina: {len(bloques)}")
        
        for i, bloque in enumerate(bloques):
            try:
                # EXTRAER NOMBRE
                nombre = "Sin nombre"
                for selector in ["h3", "h2", ".product-name", ".title", "[class*='name']"]:
                    try:
                        elem = bloque.find_element(By.CSS_SELECTOR, selector)
                        nombre = elem.text.strip()
                        if len(nombre) > 5:
                            break
                    except:
                        continue
                
                # EXTRAER MARCA (Regex mejorado para tildes)
                marca = "Sin marca"
                if nombre != "Sin nombre":
                    match = re.match(r'^([A-ZÀ-Ú][a-zà-ú]+(?:\s[A-ZÀ-Ú][a-zà-ú]+)?)', nombre)
                    if match:
                        marca = match.group(1)
                
                # EXTRAER PRECIO
                precio_elems = bloque.find_elements(By.XPATH, ".//*[contains(text(), '$')]")
                precio = 0
                
                for p_elem in precio_elems:
                    p_texto = p_elem.text.strip()
                    # Ignorar precios por unidad de medida (PUM)
                    if any(unit in p_texto.lower() for unit in ['x un', 'x kg', 'x lt', 'x ml']):
                        continue
                    
                    p_limpio = re.sub(r'[^\d]', '', p_texto)
                    if len(p_limpio) >= 3:
                        precio = int(p_limpio)
                        break
                
                # GUARDAR SI ES VALIDO
                if nombre != "Sin nombre" and precio > 0:
                    datos_finales.append({
                        "nombre_producto": nombre,
                        "precio": precio,
                        "supermercado": "Santa Isabel",
                        "categoria": "Lacteos Huevos Congelados",
                        "marca": marca,
                        "pagina": pagina,
                        "fecha_scraping": time.strftime("%Y-%m-%d %H:%M:%S")
                    })
                    print(f"{len(datos_finales):4d} | ${precio:>8,} | {marca:>12} | {nombre[:50]}")

            except Exception as e:
                continue # Si falla un bloque, saltar al siguiente

        print(f"PAGINA {pagina} FINALIZADA.")

except Exception as e:
    print(f"Error crítico en el proceso: {e}")

finally:
    if driver:
        driver.quit()
        print("\nNavegador cerrado.")

# 3. Procesamiento de datos y Reporte Final
if datos_finales:
    # Calcular precio promedio de lo capturado
    precios = [p["precio"] for p in datos_finales]
    precio_promedio_categoria = int(round(sum(precios) / len(precios)))

    # Inyectar el promedio en cada documento (denormalización útil para analítica)
    for doc in datos_finales:
        doc["precio_promedio"] = precio_promedio_categoria

    print("\n" + "="*100)
    print(f"RESUMEN DEL SCRAPING")
    print("="*100)
    print(f"Total productos capturados: {len(datos_finales)}")
    print(f"Precio promedio calculado:  ${precio_promedio_categoria:,}")
    print("-" * 100)

    # Mostrar lista detallada
    for i, producto in enumerate(datos_finales, 1):
        print(f"{i:4d}. ${producto['precio']:>8,} | {producto['marca']:>15} | {producto['nombre_producto']}")
    
    print(f"\nPROCESO COMPLETADO EXITOSAMENTE")
else:
    print("\nNo se pudieron obtener datos. Verifica la conexión o los selectores CSS.")
```
