```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, regexp_replace
# Importas tu scraper y el de tus compañeros (ejemplo: scraper_lider)
from scrapers import scraper_unimarc, scraper_lider [cite: 1506]

# 1. Recolectamos las listas de todos los integrantes
print("🚀 Iniciando integración masiva para LosAveMayo...")
data_unimarc = scraper_unimarc.ejecutar_extraccion()
data_lider = scraper_lider.ejecutar_extraccion()

# 2. Iniciamos Spark configurado para MongoDB Atlas [cite: 1518]
spark = SparkSession.builder \
    .appName("Integradora_LosAveMayo") \
    .config("spark.mongodb.output.uri", "mongodb+srv://tu_usuario:tu_pass@cluster0.xxxx.mongodb.net/BigData_UCN?retryWrites=true&w=majority") \
    .getOrCreate() [cite: 1520-1522]

# 3. Convertimos las listas en DataFrames unificados [cite: 1523]
df_unimarc = spark.createDataFrame(data_unimarc)
df_lider = spark.createDataFrame(data_lider)

# 4. UNIÓN: Spark une las tablas porque tienen los mismos nombres de columna [cite: 1529, 1560]
df_final = df_unimarc.union(df_lider)

# 5. TRANSFORMACIÓN: Limpieza masiva del precio [cite: 1530, 1535]
df_limpio = df_final.withColumn("precio", 
    regexp_replace(col("precio"), "[^0-9]", "").cast("float"))

# 6. CARGA: Guardamos en la nube Atlas de un solo golpe [cite: 1540-1541]
df_limpio.write.format("mongodb").mode("append").save()

print("✅ Datos de LosAveMayo enviados a MongoDB Atlas con éxito.")
```
