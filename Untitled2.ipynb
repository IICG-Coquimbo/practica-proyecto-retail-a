```python
from pyspark.sql import SparkSession

# 1. Cargar datos con el conector compatible para Spark 3.5
spark = SparkSession.builder \
    .appName("Analisis_Categorias") \
    .config("spark.mongodb.read.connection.uri", "mongodb://database:27017/proyecto_bigdata.datos_scraping") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.3.0") \
    .getOrCreate()

df = spark.read.format("mongodb").load()
```

```python
df.printSchema()
```

```python
df.show(5)
```

```python
from pyspark.sql.functions import col

# Cambiamos "item" por "identificador" que es la columna que Spark sí reconoce
df_filtrado = df.filter((col("identificador").isNotNull()) & (col("valor") > 0))

print("PASO 2: Limpieza completada.")
print("Registros originales:", df.count())
print("Registros validos:", df_filtrado.count())
```

```python
from pyspark.sql import functions as F

# Verificar datos
print("Columnas:", df.columns)
df.show(3)

# Análisis correcto
reporte_grupos = df.groupBy("grupo").agg(
    F.count("*").alias("cantidad_items"),
    F.avg("valor").alias("precio_promedio"),
    F.min("valor").alias("precio_minimo"),
    F.max("valor").alias("precio_maximo"),
    F.sum("valor").alias("total_ventas")
).orderBy(F.desc("precio_promedio"))

print("\n📊 REPORTES POR GRUPO:")
reporte_grupos.show()
```

```python
from pyspark.sql import functions as F

print(" VERIFICANDO GRUPOS DISPONIBLES...")
df.select("grupo").distinct().show(20, truncate=False)

# Filtrar específicamente grupo "Losavemayo"
NOMBRE_GRUPO = "Losavemayo"

print(f"\nANALIZANDO GRUPO: '{NOMBRE_GRUPO}'")

# Verificar si existe el grupo
existe_grupo = df.filter(F.col("grupo") == NOMBRE_GRUPO).count()
if existe_grupo == 0:
    print(f" ADVERTENCIA: No se encontró grupo '{NOMBRE_GRUPO}'")
    print("Grupos disponibles mostrados arriba")
else:
    print(f"Grupo '{NOMBRE_GRUPO}' encontrado: {existe_grupo:,} registros")

# TICKET DE SALIDA COMPLETO
ticket_salida = df.filter(F.col("grupo") == NOMBRE_GRUPO) \
    .agg(
        F.count("identificador").alias("total_items"),
        F.round(F.avg("valor"), 2).alias("precio_medio"),
        F.min("valor").alias("precio_minimo"),
        F.max("valor").alias("precio_maximo"),
        F.sum("valor").alias("total_ventas"),
        F.max("fecha_captura").alias("ultima_sincronizacion")
    )

print(f"\n TICKET DE SALIDA: {NOMBRE_GRUPO}")
print("=" * 50)
ticket_salida.show(truncate=False)

# RESUMEN ADICIONAL
print(f"\n📈 RESUMEN RÁPIDO:")
print(f"   • Total ítems: {ticket_salida.collect()[0]['total_items']:,}")
print(f"   • Precio medio: ${ticket_salida.collect()[0]['precio_medio']}")
print(f"   • Rango precios: ${ticket_salida.collect()[0]['precio_minimo']} - ${ticket_salida.collect()[0]['precio_maximo']}")
print(f"   • Última sincronización: {ticket_salida.collect()[0]['ultima_sincronizacion']}")
```
